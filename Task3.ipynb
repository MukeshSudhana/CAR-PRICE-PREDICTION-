```python
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df_numeric = df.drop(non_numeric_cols, axis=1)
correlation_matrix = df_numeric.corr()
```

```python
plt.figure(figsize=(16, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
